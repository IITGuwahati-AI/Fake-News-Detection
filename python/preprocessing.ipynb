```
# clean up 
# 2nd argument of replace must be str
def clean_up(label):
    if(label=='clickbait'):
      label1 = 0
    else:
      label1 = 1
    return label1
```

```
l = []
for i in data['truthClass']:
  l.append(clean_up(i))
```

```
data["label"] = l
```

```
l = []
for i in Tdata['truthClass']:
  l.append(clean_up(i))
```

```
Tdata["label"] = l
```
